```python
#SEBA
# Variables de inter√©s (en may√∫sculas)
vars_interes = ["CODUSU","NRO_HOGAR","COMPONENTE","CH04", "CH06", "CH07", "CH08", "ITF",
                "NIVEL_ED", "ESTADO", "CAT_INAC", "IPCF", "P21", "P47T", "PP3E_TOT", "PP04A","PP03D", "DECIFR", "CH09"]

# --- Base 2005 ---
df2005["ANIO"] = 2005
df2005_f = df2005[df2005["REGION"] == "Pampeana"][vars_interes + ["REGION", "ANIO"]]

# --- Base 2025 ---
df2025["ANIO"] = 2025
df2025_f = df2025[df2025["REGION"] == 41][vars_interes + ["REGION", "ANIO"]]
```

```python
# HOMOLOGACI√ìN DE LAS VARIABLES ENTRE AMBAS BASES, TANTO EN VALORES COMO EN FORMATO

# Ajustes a la base de 1T 2005
df2005_f['NRO_HOGAR'] = df2005_f['NRO_HOGAR'].astype('int64')
df2005_f['COMPONENTE'] = df2005_f['COMPONENTE'].replace({'Servicio dom√©stico en hogares': 51, 'Pensionistas en hogares': 71}).astype('int64')
df2005_f['CH04'] = df2005_f['CH04'].replace({'Var√≥n': 1, 'Mujer': 2}).astype('int64')
df2005_f['CH06'] = df2005_f['CH06'].replace({'Menos de 1 a√±o': -1, '98 y m√°s a√±os': 2}).astype('int64')   # Hay que agrupar en las dos bases las edades que son menores a 1 y los que son mde 98 o m√°s en mismo grupo
df2005_f['CH07'] = df2005_f['CH07'].replace({'Unido': 1, 'Casado': 2, 'Separado o divorciado': 3, 'Viudo': 4, 'Soltero': 5, 'Ns./Nr.': 99}).astype('int64')
df2005_f['CH08'] = df2005_f['CH08'].replace({'Obra social (incluye PAMI)': 1, 'Mutual/Prepaga/Servicio de emergencia': 2, 'Planes y seguros p√∫blicos': 3, 'No paga ni le descuentan': 4, 'Obra social y mutual/prepaga/servicio de emergencia': 12, 'Obra social y planes y seguros p√∫blicos': 13, 'Mutual/prepaga/servicio de emergencia/planes y seguros p√∫bli': 23, 'Obra social, mutual/prepaga/servicio de emergencia y planes': 123, 'Ns./Nr.': 9}).astype('int64')
df2005_f['NIVEL_ED'] = df2005_f['NIVEL_ED'].replace({'Primaria Incompleta (incluye educaci√≥n especial)': 1, 'Primaria Completa': 2, 'Secundaria Incompleta': 3, 'Secundaria Completa': 4, 'Superior Universitaria Incompleta': 5, 'Superior Universitaria Completa': 6, 'Sin instrucci√≥n': 7}).astype('int64')
df2005_f['ESTADO'] = df2005_f['ESTADO'].replace({'Entrevista individual no realizada (no respuesta al cuestion': 0, 'Ocupado': 1, 'Desocupado': 2, 'Inactivo': 3, 'Menor de 10 a√±os': 4}).astype('int64')
df2005_f['CAT_INAC'] = df2005_f['CAT_INAC'].replace({'Jubilado/pensionado': 1, 'Rentista': 2, 'Estudiante': 3, 'Ama de casa': 4, 'Menor de 6 a√±os': 5, 'Discapacitado': 6, 'Otros': 7}).astype('int64')
df2005_f['PP04A'] = df2005_f['PP04A'].replace({'...estatal?': 1, '...privada?': 2, '...de otro tipo? (especificar)': 3, 'Ns./Nr.': 99}).astype('float64')
df2005_f['DECIFR'] = df2005_f['DECIFR'].astype('int64')
df2005_f['CH09'] = df2005_f['CH09'].replace({'S√≠': 1, 'No': 2, 'Menor de 2 a√±os': 3, 'Ns./Nr.': 99}).astype('int64')

# Ajustes a la base de 1T 2025
df2025_f['CH06'] = df2025_f['CH06'].apply(lambda x: 98 if x >= 98 else x)
df2025_f['CH08'] = df2025_f['CH08'].astype('int64')
df2025_f['IPCF'] = df2025_f['IPCF'].str.replace(',', '.').astype('float64')
df2025_f['ITF'] = df2025_f['ITF'].astype('float64')
df2025_f['P21'] = df2025_f['P21'].astype('float64')
df2025_f['REGION'] = df2025_f['REGION'].replace({41: 'Pampeana'}).astype('category')
```

```python
# --- Unir bases ---
df = pd.concat([df2005_f, df2025_f], ignore_index=True)

print("‚úÖ Bases unidas y filtradas (Regi√≥n Pampeana)")
display(df.head())
print("Shape final:", df.shape)
```

```python
# --- Tu c√≥digo base ---
counts = df.groupby(['ANIO', 'SEXO']).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

colors = {}
for col in percentages.columns:
    if col.upper() == 'VARON':
        colors[col] = 'blue'
    elif col.upper() == 'MUJER':
        colors[col] = 'pink'
    else:
        colors[col] = 'gray'

# --- Crear gr√°fico ---
ax = percentages.plot(kind='bar', stacked=True, color=[colors[col] for col in percentages.columns], figsize=(8,5))

plt.title('Distribuci√≥n por sexo\nRegi√≥n Pampeana')
plt.xlabel('A√±o')
plt.ylabel('Participaci√≥n porcentual (%)')
plt.legend(title='Sexo', loc='upper center',bbox_to_anchor=(0.5, 0.95), ncol=1)
plt.xticks(rotation=0)
plt.tight_layout()

# --- Agregar porcentajes dentro de las barras ---
for i, (index, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for col in percentages.columns:
        value = row[col]
        if value > 3:  # solo mostrar si el segmento es >3% para que no se solape
            plt.text(
                i,
                cumulative + value/2,
                f"{value:.1f}%",
                ha='center', va='center', fontsize=9, color='white', fontweight='bold'
            )
        cumulative += value

plt.show()
```

```python
# --- Variables de inter√©s ---
vars_corr = ['CH04', 'CH06', 'CH07', 'CH08', 'NIVEL_ED', 'ESTADO', 'CAT_INAC', 'IPCF']

# --- Diccionario de etiquetas ---
labels = {
    'CH04': 'Sexo',
    'CH06': 'Edad',
    'CH07': 'Estado civil',
    'CH08': 'Cobertura m√©dica',
    'NIVEL_ED': 'Nivel educativo alcanzado',
    'ESTADO': 'Condici√≥n de actividad',
    'CAT_INAC': 'Categor√≠a de inactividad',
    'IPCF': 'Ingreso per c√°pita familiar'
}

# --- Filtrar y preparar los datos ---
df_corr = df[vars_corr].copy()

# Convertir categ√≥ricas a num√©ricas
df_corr = df_corr.apply(lambda x: pd.factorize(x)[0] if x.dtype == 'object' else x)

# --- Calcular la matriz de correlaci√≥n ---
corr_matrix = df_corr.corr()

# --- Renombrar filas y columnas con las etiquetas descriptivas ---
corr_matrix.rename(index=labels, columns=labels, inplace=True)

# --- Mostrar tabla num√©rica ---
print("üìä Matriz de correlaci√≥n:")
print(corr_matrix.round(2))

# --- Graficar mapa de calor ---
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Matriz de correlaci√≥n entre variables seleccionadas")
plt.tight_layout()
plt.show()

```

```python
# Valores de adulto equivalente de cada persona seg√∫n su sexo y edad
# Convertir CH06 a num√©rico (si hay errores, poner NaN)
df["CH06"] = pd.to_numeric(df["CH06"], errors="coerce")

# Asegurarse de que CH04 tambi√©n sea num√©rico
df["CH04"] = pd.to_numeric(df["CH04"], errors="coerce")

# Funci√≥n para asignar factor seg√∫n sexo y edad
def asignar_factor(sexo, edad):
    if pd.isnull(edad) or pd.isnull(sexo):
        return np.nan

    tabla = [
        (lambda e: e < 1, 0.35, 0.35),
        (lambda e: e == 1, 0.37, 0.37),
        (lambda e: e == 2, 0.46, 0.46),
        (lambda e: e == 3, 0.51, 0.51),
        (lambda e: e == 4, 0.55, 0.55),
        (lambda e: e == 5, 0.60, 0.60),
        (lambda e: e == 6, 0.64, 0.64),
        (lambda e: e == 7, 0.66, 0.66),
        (lambda e: e == 8, 0.68, 0.68),
        (lambda e: e == 9, 0.69, 0.69),
        (lambda e: e == 10, 0.70, 0.79),
        (lambda e: e == 11, 0.72, 0.82),
        (lambda e: e == 12, 0.74, 0.85),
        (lambda e: e == 13, 0.76, 0.90),
        (lambda e: e == 14, 0.76, 0.96),
        (lambda e: e == 15, 0.77, 1.00),
        (lambda e: e == 16, 0.77, 1.03),
        (lambda e: e == 17, 0.77, 1.04),
        (lambda e: 18 <= e <= 29, 0.76, 1.02),
        (lambda e: 30 <= e <= 45, 0.77, 1.00),
        (lambda e: 46 <= e <= 60, 0.76, 1.00),
        (lambda e: 61 <= e <= 75, 0.67, 0.83),
        (lambda e: e > 75, 0.63, 0.74)
    ]

    for condicion, mujer, varon in tabla:
        if condicion(edad):
            return mujer if sexo == 2 else varon
    return np.nan

# Crear la columna
df["adult_equiv"] = df.apply(lambda x: asignar_factor(x["CH04"], x["CH06"]), axis=1)

# Verificar
print(df[["CH04", "CH06", "adult_equiv"]].head(15))
```

```python
# Crear una copia segura del DataFrame filtrado
respondieron = df[df['ITF'] != 0].copy()

# Agregar la columna 'ingreso_necesario' seg√∫n el valor de ANIO
respondieron.loc[:, 'ingreso_necesario'] = respondieron.apply(
    lambda row: row['ad_equiv_hogar'] * 365177 if row['ANIO'] == 2025 else row['ad_equiv_hogar'] * 205.07,
    axis=1
)

respondieron['pobre'] = (respondieron['ITF'] < respondieron['ingreso_necesario']).astype(int)
```

```python
# Calcular cantidad total por a√±o
total_por_anio = respondieron.groupby('ANIO').size()

# Calcular cantidad de pobres por a√±o
pobres_por_anio = respondieron[respondieron['pobre'] == 1].groupby('ANIO').size()

# Calcular porcentaje de pobreza por a√±o
porcentaje_pobreza = (pobres_por_anio / total_por_anio * 100).round(2)

print("Cantidad de pobres por a√±o:")
print(pobres_por_anio)

print("Porcentaje de pobreza por a√±o:")
print(porcentaje_pobreza)
```

```python
# Labels para NIVEL_ED
nivel_ed_labels = {
    1: "Primario incompleto (incluye educaci√≥n especial)",
    2: "Primario completo",
    3: "Secundario incompleto",
    4: "Secundario completo",
    5: "Superior y universitario incompleto",
    6: "Superior y universitario completo",
    7: "Sin instrucci√≥n"
}

# Reemplazar c√≥digos por labels
respondieron['NIVEL_ED_LBL'] = respondieron['NIVEL_ED'].map(nivel_ed_labels)

# Tabla cruzada absoluta por a√±o
tabla_abs = pd.crosstab(
    [respondieron['ANIO'], respondieron['NIVEL_ED_LBL']],
    respondieron['pobre'],
    rownames=['A√±o', 'Nivel educativo'],
    colnames=['Pobre'],
    dropna=False
).rename(columns={0: "No pobre", 1: "Pobre"})

# Tabla cruzada en porcentaje por a√±o
tabla_pct = tabla_abs.div(tabla_abs.sum(axis=1), axis=0) * 100
tabla_pct = tabla_pct.round(2)

# Mostrar resultados
print("üìä Tabla cruzada absoluta por a√±o:")
display(tabla_abs)

print("\nüìä Tabla cruzada en porcentaje por a√±o:")
display(tabla_pct)
```
