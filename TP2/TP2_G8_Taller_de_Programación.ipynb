```python
# Valores de adulto equivalente de cada persona seg√∫n su sexo y edad
# Convertir CH06 a num√©rico (si hay errores, poner NaN)
df["CH06"] = pd.to_numeric(df["CH06"], errors="coerce")

# Asegurarse de que CH04 tambi√©n sea num√©rico
df["CH04"] = pd.to_numeric(df["CH04"], errors="coerce")

# Funci√≥n para asignar factor seg√∫n sexo y edad
def asignar_factor(sexo, edad):
    if pd.isnull(edad) or pd.isnull(sexo):
        return np.nan

    tabla = [
        (lambda e: e < 1, 0.35, 0.35),
        (lambda e: e == 1, 0.37, 0.37),
        (lambda e: e == 2, 0.46, 0.46),
        (lambda e: e == 3, 0.51, 0.51),
        (lambda e: e == 4, 0.55, 0.55),
        (lambda e: e == 5, 0.60, 0.60),
        (lambda e: e == 6, 0.64, 0.64),
        (lambda e: e == 7, 0.66, 0.66),
        (lambda e: e == 8, 0.68, 0.68),
        (lambda e: e == 9, 0.69, 0.69),
        (lambda e: e == 10, 0.70, 0.79),
        (lambda e: e == 11, 0.72, 0.82),
        (lambda e: e == 12, 0.74, 0.85),
        (lambda e: e == 13, 0.76, 0.90),
        (lambda e: e == 14, 0.76, 0.96),
        (lambda e: e == 15, 0.77, 1.00),
        (lambda e: e == 16, 0.77, 1.03),
        (lambda e: e == 17, 0.77, 1.04),
        (lambda e: 18 <= e <= 29, 0.76, 1.02),
        (lambda e: 30 <= e <= 45, 0.77, 1.00),
        (lambda e: 46 <= e <= 60, 0.76, 1.00),
        (lambda e: 61 <= e <= 75, 0.67, 0.83),
        (lambda e: e > 75, 0.63, 0.74)
    ]

    for condicion, mujer, varon in tabla:
        if condicion(edad):
            return mujer if sexo == 2 else varon
    return np.nan

# Crear la columna
df["adult_equiv"] = df.apply(lambda x: asignar_factor(x["CH04"], x["CH06"]), axis=1)

# Verificar
print(df[["CH04", "CH06", "adult_equiv"]].head(15))
```

```python
import pandas as pd

# --- 1Ô∏è‚É£ Unificar nombres de columnas ---
df.columns = df.columns.str.upper()

# --- 2Ô∏è‚É£ Verificar existencia de las variables requeridas ---
if "POBRE" not in df.columns:
    raise ValueError("‚ö†Ô∏è No se encontr√≥ la variable 'POBRE' en el DataFrame. Verifica el nombre exacto.")
if "ANIO" not in df.columns:
    raise ValueError("‚ö†Ô∏è No se encontr√≥ la variable 'A√ëO' en el DataFrame. Es necesaria para comparar 2005 y 2025.")

# --- 3Ô∏è‚É£ Definir a√±os de inter√©s ---
anios = [2005, 2025]
resumen = {}

for anio in anios:
    df_anio = df[df["ANIO"] == anio]

    total_obs = len(df_anio)
    na_pobre = df_anio["POBRE"].isna().sum()
    cant_pobres = (df_anio["POBRE"] == 1).sum()
    cant_no_pobres = (df_anio["POBRE"] == 0).sum()
    vars_limpias = df_anio.columns[df_anio.notna().all()].shape[0]

    resumen[anio] = [total_obs, na_pobre, cant_pobres, cant_no_pobres, vars_limpias]

# --- 4Ô∏è‚É£ Crear tabla resumen comparativa ---
tabla_resumen = pd.DataFrame({
    "Indicador": [
        "Cantidad total de observaciones",
        "Observaciones con NA en POBRE",
        "Cantidad de pobres (POBRE == 1)",
        "Cantidad de no pobres (POBRE == 0)",
        "Cantidad de variables limpias y homogeneizadas"
    ],
    "2005": resumen.get(2005, [None]*5),
    "2025": resumen.get(2025, [None]*5)
})

# --- 5Ô∏è‚É£ Mostrar resultados ---
print("\nüìä Tabla comparativa por a√±o (2005 vs 2025):\n")
print(tabla_resumen.to_string(index=False))
```

```python
# JORGE
# ============================================
# PCA con seis variables seleccionadas (con manejo de NaN)
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# -----------------------------
# 1) Selecci√≥n de variables
# -----------------------------
variables = ["CH06", "EDAD2", "AEDUC", "ingreso_total_familiar", "MIEMBROS", "HORASTRAB"]
X = df[variables].copy()

# Renombrar columnas para claridad
X.rename(columns={
    "CH06": "edad",
    "EDAD2": "edad2",
    "AEDUC": "educ",
    "MIEMBROS": "miembros_hogar",
    "HORASTRAB": "horastrab"
}, inplace=True)

# -----------------------------
# 2) Estandarizar variables
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 3) Imputar valores faltantes (NaN) por la media
# -----------------------------
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(X_scaled)

# -----------------------------
# 4) Aplicar PCA
# -----------------------------
pca = PCA(n_components=6)
scores = pca.fit_transform(X_imputed)   # Coordenadas de observaciones
loadings = pca.components_.T            # Ponderadores de variables
explained_var = pca.explained_variance_ratio_

# -----------------------------
# 5) Gr√°fico de dispersi√≥n: Componente 1 vs Componente 2
# -----------------------------
plt.figure(figsize=(7,5))
plt.scatter(scores[:,0], scores[:,1], alpha=0.5)
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.title("PCA: Scores del primer y segundo componente")
plt.grid(True)
plt.show()

# -----------------------------
# 6) Gr√°fico de loadings (ponderadores)
# -----------------------------
plt.figure(figsize=(7,7))
for i, var in enumerate(X.columns):
    plt.arrow(0, 0, loadings[i,0], loadings[i,1],
              color='r', alpha=0.7, head_width=0.05, head_length=0.05)
    plt.text(loadings[i,0]*1.15, loadings[i,1]*1.15, var, color='b', ha='center', va='center')

plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.title("PCA: Loadings del primer y segundo componente")
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.grid()
plt.show()

# -----------------------------
# 7) Gr√°fico de varianza explicada
# -----------------------------
plt.figure(figsize=(7,5))
plt.bar(range(1,7), explained_var*100, color='skyblue', edgecolor='black', label='Varianza por componente')
plt.plot(range(1,7), np.cumsum(explained_var)*100, marker='o', color='red', label='Varianza acumulada')
plt.xticks(range(1,7))
plt.ylabel("Varianza explicada (%)")
plt.xlabel("Componente")
plt.title("PCA: Varianza explicada por componente")
plt.legend()
plt.show()

# -----------------------------
# 8) Comentarios
# -----------------------------
print("""
Comentarios :
- El gr√°fico de scores muestra c√≥mo se distribuyen los individuos seg√∫n los dos primeros componentes principales.
- Los loadings indican qu√© variables contribuyen m√°s a cada componente. Por ejemplo, si 'edad2' tiene un peso alto en el primer componente, la edad influye fuertemente en esa dimensi√≥n.
- La varianza explicada indica qu√© tan representativos son los componentes. La l√≠nea roja de varianza acumulada permite decidir cu√°ntos componentes retener para un an√°lisis resumido.
""")
```

```python
#JORGE
# ============================================
# K-means clustering: edad vs ingreso_total_familiar
# ============================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# -----------------------------
# 1Ô∏è‚É£ Selecci√≥n de variables para clustering
# -----------------------------
X_cluster = df[["CH06", "ingreso_total_familiar"]].copy()
X_cluster.rename(columns={"CH06":"edad"}, inplace=True)

# Estandarizar variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# -----------------------------
# 2Ô∏è‚É£ Funci√≥n para correr K-means y graficar clusters
# -----------------------------
def plot_kmeans(X_scaled, k, df_plot):
    kmeans = KMeans(n_clusters=k, n_init=20, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)

    plt.figure(figsize=(7,5))
    sns.scatterplot(
        x=df_plot["edad"],
        y=df_plot["ingreso_total_familiar"],
        hue=clusters,
        palette="Set2",
        alpha=0.6
    )
    plt.xlabel("Edad")
    plt.ylabel("Ingreso total familiar")
    plt.title(f"K-means clustering con k={k}")
    plt.legend(title="Cluster")
    plt.show()

    return kmeans.inertia_

# -----------------------------
# 3Ô∏è‚É£ Correr K-means con k=2,4,10
# -----------------------------
inertias_selected = []
for k in [2,4,10]:
    inertia = plot_kmeans(X_scaled, k, X_cluster)
    inertias_selected.append((k,inertia))

# -----------------------------
# 4Ô∏è‚É£ Medida de disimilitud para k=1 a k=10 (Elbow method)
# -----------------------------
inertia_list = []
k_range = range(1,11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=20, random_state=42)
    kmeans.fit(X_scaled)
    inertia_list.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
plt.plot(k_range, inertia_list, marker='o', linestyle='--', color='b')
plt.xlabel("N√∫mero de clusters (k)")
plt.ylabel("Inercia (suma de distancias al cuadrado)")
plt.title("Elbow method para determinar n√∫mero √≥ptimo de clusters")
plt.xticks(k_range)
plt.grid(True)
plt.show()

# -----------------------------
# 5Ô∏è‚É£ Comparaci√≥n visual con pobreza
# -----------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(
    x=X_cluster["edad"],
    y=X_cluster["ingreso_total_familiar"],
    hue=df["POBRE"],  # 0 = no pobre, 1 = pobre
    palette={0:"green", 1:"red"},
    alpha=0.6
)
plt.xlabel("Edad")
plt.ylabel("Ingreso total familiar")
plt.title("Distribuci√≥n de individuos por condici√≥n de pobreza")
plt.legend(title="Pobreza")
plt.show()

# -----------------------------
# 6Ô∏è‚É£ Comentarios sugeridos
# -----------------------------
print("""
Comentarios sugeridos:
- Con k=2, los clusters no separan perfectamente a los pobres y no pobres, aunque puede reflejar parcialmente la condici√≥n socioecon√≥mica.
- Con k=4 o k=10 se observan subgrupos de ingreso y edad que podr√≠an reflejar distintos niveles socioecon√≥micos.
- El gr√°fico de Elbow indica el n√∫mero √≥ptimo de clusters donde la disminuci√≥n de inercia empieza a estabilizarse.
- La visualizaci√≥n con colores por pobreza muestra qu√© tan bien los clusters se alinean con la condici√≥n real de pobreza.
""")
```

```python
# JORGE
# ==============================
# Cluster Jer√°rquico
# ==============================
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Selecci√≥n de variables
variables = ["CH06", "EDAD2", "AEDUC", "ingreso_total_familiar", "MIEMBROS", "HORASTRAB"]
X = df[variables].copy()

# Renombrar columnas para mayor claridad
X.rename(columns={
    "CH06": "edad",
    "EDAD2": "edad2",
    "AEDUC": "educ",
    "PP3E_TOT": "miembros_hogar",
    "HORASTRAB": "horastrab"
}, inplace=True)

# Eliminar filas con valores faltantes
X_clean = X.dropna()

# Estandarizar variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)

# Calcular linkage usando el m√©todo de Ward
Z = linkage(X_scaled, method='ward')

# Graficar dendrograma
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=10, color_threshold=5, leaf_rotation=90)
plt.title("Dendrograma - Clustering Jer√°rquico")
plt.xlabel("Observaciones (agrupadas)")
plt.ylabel("Distancia/Disimilitud")
plt.show()
```

**Interpretaci√≥n**
Un dendrograma es un diagrama en forma de √°rbol que representa la estructura de agrupamiento de un conjunto de observaciones en un an√°lisis de clustering jer√°rquico. Muestra qui√©n se agrupa con qui√©n y qu√© tan parecidos son los grupos en un an√°lisis jer√°rquico.

*   Cada hoja representa una persona.
*   Ramas bajas: individuos similares en edad, ingreso y dem√°s caracter√≠sticas.
*   Ramas altas: grupos muy distintos.

M√©todo Ward: Minimiza la varianza dentro de los clusters

* El eje vertical muestra la disimilitud entre observaciones (mayor altura ‚Üí mayor diferencia); el eje horizontal representa las observaciones agrupadas; las l√≠neas indican d√≥nde se fusionan clusters.
* Las observaciones comienzan individualmente y se agrupan progresivamente en peque√±os clusters; a medida que la distancia aumenta, todos los datos se combinan en un solo cluster.
* Cortando el dendrograma a cierta altura (por ejemplo, ~100) se pueden identificar 3-4 clusters principales, representando los grupos m√°s relevantes en los datos.
